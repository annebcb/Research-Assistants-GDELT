# Overtime Sentiment Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# This function is used to select the most frequent occuring nouns
def top_noun_selection(df):
    Topic = ''
    
    # 
    for i in range(len(df['Title'].values)):
        Topic = Topic + str(df['Title'].values[i] + " ")
    # text

    tokenizer = RegexpTokenizer(r'\w+')
    # split text into smaller unit
    words = tokenizer.tokenize(Topic)

    stop_words = stopwords.words('english')

    filter_text_1 = [w for w in words if w not in stop_words]

    lemmatizer = WordNetLemmatizer()

    filter_text_2 = []
    filter_text_3 = []
    
    # lemmatize every single word
    for i in range(0, len(filter_text_1)):
        filter_text_2.append(lemmatizer.lemmatize(filter_text_1[i]))

    for i in range(0, len(filter_text_2)):
        filter_text_3.append(lemmatizer.lemmatize(filter_text_2[i], pos = 'v'))
    
    # change everything into lower case
    def lower_word(all_content):
        for i, content in enumerate(all_content):
            all_content[i] = content.lower()
        return all_content

    filter_text_4 = lower_word(filter_text_3)

    text_1 = ''

    for i in range(0, len(filter_text_4)):
        text_1 = text_1 + filter_text_4[i]+ " "

    # function to test if something is a noun
    is_noun = lambda pos: pos[:2] == 'NN'

    tokenized = nltk.word_tokenize(text_1)
    nouns = [word for (word, pos) in nltk.pos_tag(tokenized) if is_noun(pos)] 
    
    # count the word frequency
    ds = pd.Series(nouns).value_counts()

    word_count_df = ds.to_frame()
    
    return(word_count_df)
```

```python
# Over time Analysis
def OverTimeAnalysis(df):

    neg = []
    neu = []
    pos = []
    compound = []

    sentences = list(df['Title'].values)

    for sentence in sentences:
        # SentimentIntensityAnalyzer will give you four scores of sentiment.
        # For this project, i chose the compound score.
        sid = SentimentIntensityAnalyzer()
        ss = sid.polarity_scores(sentence)
        neg.append(ss['neg'])
        neu.append(ss['neu'])
        pos.append(ss['pos'])
        compound.append(ss['compound'])
    df['compound_sentiment'] = compound
    # data processing of the date
    # For clear visualization purpose, i create the averge sentiment score based on quarters
    df['Date'] = df['Date'].astype('datetime64')
    df['Date'] = df['Date'].astype(str)

    df['Year'] = df['Date'].apply(lambda x: x[0:4])
    df['Month'] = df['Date'].apply(lambda x: x[5:7])
    df['Month'] = df['Month'].astype(int)
    df['Year'] = df['Year'].astype(int)
    
    df.loc[(df['Month'] <= 12), 'Quarter'] = 'Q4'
    df.loc[(df['Month'] <= 9), 'Quarter'] = 'Q3'
    df.loc[(df['Month'] <= 6), 'Quarter'] = 'Q2'
    df.loc[(df['Month'] <= 3), 'Quarter'] = 'Q1'
    
    df['Year'] = df['Year'].astype(str)
    df['Year-Quarter'] = df['Year'] + " " + df['Quarter']
    

    df2 = df.groupby('Year-Quarter')['compound_sentiment'].sum()  # do average sentiment based on quarters, show clearly
    df_date_sentiment = df2.to_frame()

    df_date_sentiment = df_date_sentiment.reset_index()

    plt.style.use('ggplot')

    fig = plt.figure(figsize=(20, 15))

    plt.plot(df_date_sentiment['Year-Quarter'],
             df_date_sentiment['compound_sentiment'],
             linestyle = '-',
             linewidth = 2,
             color = 'steelblue',
             marker = 'o',
             markersize = 6,
             markeredgecolor='black',
             markerfacecolor='brown')

    plt.title('Sentiment Over Time')
    plt.xlabel('Date')
    plt.ylabel('Compound Sentiment')

    plt.tick_params(top = 'off', right = 'off')

    fig.autofmt_xdate(rotation = 90)
    
    plt.savefig('overtime.svg', format = 'svg')

    plt.show()
```
